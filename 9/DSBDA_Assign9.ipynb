```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('titanic')
data.head(10)
```

```python
data['age'] = data['age'].fillna(data['age'].mean())
data.head(10)
```

```python
sns.boxplot(x = 'sex' , y = 'age' , data = data)
```

```python
sns.boxplot(x = 'age' , data = data)
```

calculating the first quartile of the 'age' variable in your dataset using NumPy's percentile function.

**Interpolation** is the process of estimating unknown values that lie between known values. In the context of calculating percentiles or quartiles from a dataset, interpolation determines how to estimate the percentile value when it falls between two observed data points. The interpolation method determines how the percentile value is calculated when it falls between two data points.

**Quartiles** are statistical measures used to divide a dataset into four equal parts, each containing approximately 25% of the data. Quartiles are often used to analyze the distribution of numerical data and to understand the spread and central tendency of a dataset.

```python
Q1 = np.percentile(data['age'] , 25 , interpolation = 'midpoint')
Q1
```

```python
Q3 = np.percentile(data['age'] , 75 , interpolation = 'midpoint')
Q3
```

**Inter Quartile Range(IQR)**

The Interquartile Range (IQR) is a measure of statistical dispersion that quantifies the spread of data points within the middle 50% of a dataset. It is calculated as the difference between the third quartile (Q3) and the first quartile (Q1).

```python
IQR = Q3 - Q1
IQR
```

```python
data.shape
```

**where** function to identify indices where the age values in your dataset are greater than or equal to a threshold defined as Q3 + 0.1 * IQR


```python
upper = np.where(data['age'] >= (Q3 + 0.1 * IQR))
upper
```

**where** function to identify indices where the age values in your dataset are lower than or equal to a threshold defined as Q3 + 0.1 * IQR

```python
lower = np.where(data['age'] <= (Q1 + 0.1 * IQR))
lower
```

This will drop rows from the DataFrame data at the indices specified in upper. Setting inplace=True ensures that the changes are made directly to the DataFrame data, and axis=0 specifies that rows should be dropped.

```python
data.drop(upper[0] , inplace = True)
```

This code will drop rows from the DataFrame data at the indices specified in lower. Setting inplace=True ensures that the changes are made directly to the DataFrame data, and axis=0 specifies that rows should be dropped.

```python
data.drop(lower[0] , inplace = True)
```
