**Defining Independent Variables**

x is an independent variable on which y variable is dependent.

'rm' is not included in x because it has null values and Linear regression do not take null values.

If you want to use 'rm' in x then first you need to remove null values with the help of various replacing techniques. for e.g. mean()

```python
x = data[['crim', 'zn',	'indus', 'chas', 'nox',	'age',	'dis',	'rad',	'tax',	'ptratio',	'b',	'lstat']]
x
```

**Defining dependent variable**

y is a dependent variable.

our model will generate y variable values based on x variable values.

```python
y = data[['medv']]
y
```

**Dividing data for model training**

train_test_split is a function which divides dataset into training and testing data.

we define 4 variables in train_test_split

x_train = training data for model of x variable

y-train = training data for model of y variable

x_train = testing data for model of x variable

y_train = testing data for model of y variable

test_size = test size defines how much data we want to give for model testing. Test size is between 0 to 1

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

**Linear Regression**

It mathematically models the dependent variable and the independent variable as a linear equation.

i.e. y = mx + c

fit() =  the fit method is the one who will determine the values of the coefficients a and b in the equation y = ax + b of the regressor, according to the training data.

Basically fit method is used for training our model

```python
lm = LinearRegression()
lm.fit(x_train, y_train)
```

predict() = predict function is used to predict values of the dependent variable based on previous training.

predict function is used in testing dataset where model performs based on its learning from training dataset.

```python
predictions = lm.predict(x_test)
```

```python
from sklearn import metrics
```

**Mean absolute error**

MAE is the average of the absolute differencebetween the predicted and actual values.

It is also known as L1 noem or Manhattan distance.

MAE gives equal weight to all errors, regardless of their magnitude. It is more robust to outliers and noise than MSE, but it can also understand the impact of larhe errors.

```python
metrics.mean_absolute_error(y_test, predictions)
```

**Mean Squared Error**

MSE is the average of the squared differences between the predicted and actual values.

It is also known as L2 norm or the Euclidean distance.

MSE gives more weight to larger errors than smaller errors. This is useful if we want our model to be sensitive to large deviations.

```python
metrics.mean_squared_error(y_test, predictions)
```

**Root Mean Squared Error**

RMSE is a square root of MSE

```python
np.sqrt(metrics.mean_squared_error(y_test, predictions))
```
