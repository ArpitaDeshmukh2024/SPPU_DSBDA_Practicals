**LogisticRegression()**

Importing LogisticRegression from scikit-learn allows us to utilize logistic regression for classification tasks. This module provides various functionalities for logistic regression, including fitting models to data, making predictions, and evaluating model performance.

**Logistic Regression**

Logistic regression is a statistical method used for binary classification tasks, where the goal is to predict the probability that an instance belongs to one of two classes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
data
```

```python
data.dtypes
```

**.iloc**: This is a method provided by pandas for selection by position. It is used to select rows and columns by their integer position in the DataFrame.

[:, [2, 3]]: This part of the code is selecting all rows (:) and the columns at positions 2 and 3 ([2, 3]).

**.values**: This attribute returns a NumPy array containing the selected values from the DataFrame.

```python
x = data.iloc[:,[2,3]].values
x
```

```python
y = data.iloc[:,4].values
y
```

imports the **train_test_split function** from the **model_selection** module of the scikit-learn library. This function is commonly used to split datasets into random train and test subsets.

**x** = independednt variables(feature values)

**y** = dependent variable(target values)

**test_size** = It specifies the proportion of the dataset to include in the test split. Here, it's set to 0.2, indicating that 20% of the data will be used for testing, while the remaining 80% will be used for training.

**random_state** = It sets the random seed for reproducibility. It ensures that the data splitting is deterministic, meaning that running the code multiple times will produce the same results.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.2, random_state = 0)
```

imports the **StandardScaler** class from the **preprocessing** module of scikit-learn.

 **StandardScaler** = it is a method for standardizing features by removing the mean and scaling to unit variance.

**fit_transform(training_variable)** = This method of the StandardScaler object is called on the training data (x_train). This method computes the mean and standard deviation of each feature in the training set and then standardizes the features based on these statistics. The standardized features are then assigned back to x_train.

**transform(testing_variable)** = This method of the StandardScaler object is called on the testing data (x_test). This method uses the mean and standard deviation computed from the training set to standardize the features of the testing set. It's important to note that the same scaling factors computed on the training set are applied to the testing set to ensure consistency.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

imports the **LogisticRegression** class from the **linear_model** module of scikit-learn

**classifier** = LogisticRegression()  :   instance of the LogisticRegression class. By default, this creates a logistic regression classifier with certain hyperparameters.

**fit(independent_training_varible , dependent_training_variable)** : This method of the LogisticRegression object is called to train the logistic regression classifier on the training data (x_train and y_train).

This method fits the logistic regression model to the training data by estimating the parameters (coefficients) that best fit the relationship between the input features (x_train) and the target variable (y_train).

After this, the logistic regression classifier (classifier) has been trained on the training data and is ready to make predictions on new data.

```python
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

**classifier.predict(x_test)**: This line uses the predict method of the trained classifier object to make predictions on the test data (x_test).

The predict method takes the test data as input and returns the predicted class labels for each instance in the test set.

```python
y_pred = classifier.predict(x_test)
y_pred
```

imports the **confusion_matrix** function from the **metrics** module of scikit-learn.

**Confusion Matrix** = It is a table that summarizes the performance of a classification model on a set of test data.

The confusion_matrix function is called with two arguments:

y_test: The true labels (actual class labels) of the test data.

y_pred: The predicted labels (predicted class labels) of the test data.

This function computes the confusion matrix based on the true and predicted labels.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix : ')
print(cm)
```

imports the **accuracy_score** function from the **metrics** module of scikit-learn.

**Accuracy Score** = It is a measure of the overall correctness of the classifier's predictions.

The accuracy_score function is called with two arguments: y_test and y_pred

This function computes the accuracy score by comparing the true labels with the predicted labels and calculating the proportion of correct predictions.

The calculated accuracy score is then multiplied by 100 to express it as a percentage.

```python
from sklearn.metrics import accuracy_score
ac = accuracy_score(y_test, y_pred) * 100
print('Accuracy Score : ')
print(ac)
```

**tp = total positive** : The count of instances where the true class is positive (1) and the classifier predicts positive (1).

**fp = false positive** : The count of instances where the true class is negative (0) but the classifier predicts positive (1).

**fn = false negative** : The count of instances where the true class is positive (1) but the classifier predicts negative (0).

**tn = true negative** : The count of instances where the true class is negative (0) and the classifier predicts negative (0).

**total** =  summing up all these values to calculate the total number of instances in the confusion matrix.

```python
tp = cm[0][0]
fp = cm[1][0]
fn = cm[0][1]
tn = cm[1][1]
total = tp + fp + fn + tn
print(total)
```

The **error rate** represents the proportion of incorrectly classified instances relative to the total number of instances. It provides a measure of how often the classifier makes mistakes in its predictions.

```python
error_rate = (fp + fn) / total
print('error rate : ')
print(error_rate)
```

imports the **classification_repor**t function from the **metrics** module of scikit-learn.

The classification report provides a summary of various classification metrics, such as precision, recall, F1-score, and support, for each class in the dataset.

The classification_report function is called with two arguments: y_test and y_pred.

Precision: The ratio of true positives to the total number of instances predicted as positive. It measures the ability of the classifier not to label a negative sample as positive.

Recall: The ratio of true positives to the total number of actual positive instances. It measures the ability of the classifier to find all the positive samples.

F1-score: The harmonic mean of precision and recall. It provides a single metric that balances precision and recall.

Support: The number of actual occurrences of each class in the test dataset.

macro avg : It computes the metrics independently for each class and then averages them.

weighted avg : It is used to compute an overall metric that takes into account the contribution of each class, with more emphasis placed on classes with a larger number of instances.

```python
from sklearn.metrics import classification_report
print('Classification Report : ')
print(classification_report(y_test, y_pred))
```
