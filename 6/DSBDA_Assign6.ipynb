```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_csv('Iris.csv')
data
```

```python
data.describe()
```

```python
x = data.iloc[:,1:5]
x
```

```python
y = data.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test ,  y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 0)
```

**GaussianNB**

imports the GaussianNB class from the naive_bayes module of scikit-learn.

Gaussian Naive Bayes is a probabilistic classifier based on the Bayes' theorem and assumes that features are independent and normally distributed within each class.

The fit method trains the Gaussian Naive Bayes classifier on the training data by estimating the parameters needed for classification. Specifically, it estimates the mean and standard deviation of each feature for each class in the training data.

```python
model = GaussianNB().fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
ac = accuracy_score(y_test, y_pred) * 100
ac
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test , y_pred)
print(cm)
```

**Error Rate for Setosa Class**

```python
tp = cm[0][0]
fn = cm[0][1] + cm[0][2]
fp = cm[1][0] + cm[2][0]
tn = cm[1][1] + cm[1][2] + cm[2][1] + cm[2][2]

total = tp + fn + fp + tn
print('True Positive : ' , tp)
print('True Negative : ' , tn)
print('False Positive : ' , fp)
print('False Negative : ' , fn)
print('Total : ' , total)
```

```python
error_rate = (fp + fn) / total
print('Error Rate : ' , error_rate)
```

**Error Rate for Versicolor Class**

```python
tp = cm[1][1]
fn = cm[1][0] + cm[1][2]
fp = cm[0][1] + cm[2][1]
tn = cm[0][0] + cm[0][2] + cm[2][0] + cm[2][2]

total = tp + fn + fp + tn
print('True Positive : ' , tp)
print('True Negative : ' , tn)
print('False Positive : ' , fp)
print('False Negative : ' , fn)
print('Total : ' , total)
```

```python
error_rate = (fp + fn) / total
print('Error Rate : ' , error_rate)
```
