```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Uploading Dataset using Seaborn**

```python
data = sns.load_dataset('titanic')
data
```

```python
data.head(10)
```

**Filling Missing Values**

```python
data['age'] = data['age'].fillna(data['age'].mean())
data.head(10)
```

**Categorical Plot(catplot)**

 Categorical plot is a type of data visualization that displays the relationship between one categorical variable and one or more other variables. It is particularly useful for visualizing the distribution of data or the relationship between categorical variables.

 Categorical plots can take various forms , such as Bar Plot , Count Plot , Point Plot , Box Plot , Violin Plot , Strip Plot , Swarm Plot

**x='sex'** specifies that the variable 'sex' will be plotted on the x-axis.

**hue='survived'** indicates that the 'survived' variable will be used to differentiate the elements in the plot. This means that the plot will display counts of 'sex' categories, but each bar will be further distinguished by whether the individuals survived or not.

**kind='count'** specifies that a count plot will be generated, showing the frequency of each category.

```python
sns.catplot(x = 'sex' , hue = 'survived' , kind = 'count' , data = data)
```

pclass_survived = group.size().unstack(): This calculates the size (count) of each group and then unstacks the result to reshape the data into a more readable format.

Specifically, it creates a new DataFrame where the index corresponds to the unique values of 'pclass' and the columns correspond to the unique values of 'survived', with the counts of occurrences in the cells.

So, pclass_survived will be a DataFrame where rows represent different passenger classes ('pclass') and columns represent survival outcomes ('survived'), and the values in the cells represent the counts of passengers belonging to each combination of class and survival status.

```python
group = data.groupby(['pclass' , 'survived'])
pclass_survived = group.size().unstack()
```

**Heatmap Visualization**

A heatmap is a graphical representation of data in a matrix format, where values in the matrix are represented as colors. Heatmaps are useful for visualizing the magnitude of relationships between two or more variables, making them particularly effective for exploring large datasets or correlation matrices.

**sns.heatmap()** : This function from Seaborn is used to create a heatmap plot.

**pclass_survived** : This is the DataFrame containing the counts of passengers grouped by passenger class and survival status, which you previously created.

**annot=True** : This parameter specifies that you want to annotate the heatmap with the numeric values from the DataFrame.

**fmt='d'** : This parameter specifies the format of the annotations. In this case, 'd' indicates that the annotations should be formatted as integers.

**fmt='.2f'** : This specifies that the annotations should be formatted as floating-point numbers with two decimal places. Adjust the number after the decimal point (2 in this case) to change the precision of the floating-point numbers in the annotations.

```python
sns.heatmap(pclass_survived , annot = True , fmt = 'd' )
```

```python
sns.heatmap(pclass_survived , annot = False , fmt = 'd' )
```

**Violine Plot**

A combination of a box plot and a kernel density plot, showing the distribution of data.

It shows the distribution of quantitative data across several levels of one or more categorical variables.

**split=True** : This parameter is used to split the violins for each category of 'sex' based on the 'survived' variable. One half of the violin will represent individuals who survived, and the other half will represent individuals who did not survive.

The resulting violin plot will show the distribution of ages for males and females, with violins split by survival outcome. This can help visualize how age distributions vary between genders and how they differ for survivors and non-survivors.

```python
sns.violinplot(x = 'sex' , y = 'age' , hue = 'survived' , data = data , split = True)
```

```python
sns.violinplot(x = 'sex' , y = 'age' , hue = 'survived' , data = data , split = False)
```

**Distribution plot**

A distribution plot, also known as a histogram, is a graphical representation of the frequency distribution of a dataset. It displays the distribution of numerical data by dividing the data into intervals (or bins) and plotting the number of observations that fall into each interval.

**data['fare']** : This specifies the 'fare' column from the dataset data, indicating the numerical variable whose distribution you want to visualize.

**kde=False** : This parameter specifies that you do not want to overlay a kernel density estimate (KDE) on the histogram.

When **kde=True**, a smoothed line representing the KDE is overlaid on the histogram; setting it to False removes this overlay.

**Kernel Density Estimation (KDE)** is a technique used for estimating the probability density function (PDF) of a continuous random variable. It provides a non-parametric way to estimate the probability density function of a dataset, which can be useful for visualizing the underlying distribution of the data or for performing density-based statistical analysis.

```python
sns.displot(data['fare'] , kde = False)
```

```python
sns.displot(data['fare'] , kde = True)
```

**Strip Plot**

This plot draws a scatter plot where one variable is categorical.

Strip plot is a type of categorical scatter plot that displays individual data points along one categorical axis and one numerical axis. It is particularly useful for visualizing the distribution of a continuous variable within different categories.

**jitter=False** : This parameter controls whether jitter (random noise) should be added to the categorical axis. Setting it to False means that data points will be plotted in a straight line without any random displacement along the categorical axis.

**jitter=True** : This parameter controls whether jitter (random noise) should be added to the categorical axis. Setting it to True means that a small amount of random displacement will be added to the data points along the categorical axis, helping to spread them out and prevent overlapping.

```python
sns.stripplot(x = 'sex' , y = 'age' , data = data , jitter = False)
```

```python
sns.stripplot(x = 'sex' , y = 'age' , data = data , jitter = True)
```

**Box Plot**

A box plot, also known as a box-and-whisker plot, it displays the distribution of data based on quartiles, showing the median, upper and lower quartiles, and outliers.

Box plots are commonly used for comparing distributions of numerical data across different categories or groups, making them valuable tools in exploratory data analysis and statistical inference. They allow quick visualization of the spread and central tendency of the data, as well as detection of any potential outliers.

```python
sns.boxplot(x = 'sex' , y = 'age' , data = data , hue = 'survived')
```

**Bar Plot**

A bar plot is a graphical representation of categorical data with rectangular bars. Bar plots are commonly used in data visualization for exploratory data analysis, comparison of groups or categories, and presentation of summary statistics. They are widely used in fields such as statistics, business, social sciences, and data science for communicating insights from categorical data.

**estimator='std'** : This parameter specifies the estimator function to compute the height of the bars. In this case, 'std' indicates that the standard deviation of ages will be computed for each category of 'sex', and the resulting values will be plotted as the heights of the bars.

The resulting bar plot will display the standard deviation of ages for males and females. Each bar will represent the standard deviation of ages within each gender group, providing insights into the variability of ages within each category.

```python
sns.barplot(x = 'sex' , y = 'age' , data = data , estimator = 'std')
```

**Count Plot**

Similar to a bar plot, but specifically designed for counting the occurrences of each category.

Count plots are widely used in data visualization for summarizing the distribution of categorical variables and comparing the frequencies of different categories. They are particularly useful for understanding the composition of categorical variables in a dataset.

```python
sns.countplot(x = 'sex' , data = data)
```
